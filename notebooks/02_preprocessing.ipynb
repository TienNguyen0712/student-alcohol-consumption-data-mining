## **üîó Data Merging (G·ªôp b·∫£ng)**

```python
merge_columms = ["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"]

student_mat_por = pd.merge(student_mat, student_por, on=merge_columms
                                                  , how="outer")
```

## **üîÑ Fixing Data Types (S·ª≠a ki·ªÉu d·ªØ li·ªáu)**

```python
student_mat_por["G_mean"] =  student_mat_por[
    ["G1_mat","G2_mat","G3_mat","G1_por","G2_por","G3_por"]].mean(axis=1, skipna=True)
```

```python
student_mat_por["Dalc"] = student_mat_por["Dalc_x"].combine_first(student_mat_por["Dalc_y"])
student_mat_por = student_mat_por.drop(columns=["Dalc_x", "Dalc_y"])
```

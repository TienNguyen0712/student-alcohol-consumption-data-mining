```python
import pandas as pd
import numpy as np
```

```python
student_mat = pd.read_csv('student-mat.csv')
student_por = pd.read_csv('student-por.csv')
```

```python
student_mat.rename(columns={'G1':'G1_mat', 'G2':'G2_mat', 'G3':'G3_mat'}, inplace=True) # ƒê·ªïi t√™n cac c·ªôt khi g·ªôp d·ªÖ ph√¢n bi·ªát
student_por.rename(columns={'G1':'G1_por', 'G2':'G2_por', 'G3':'G3_por'}, inplace=True)
```

## **üîó Data Merging (G·ªôp b·∫£ng)**

```python
merge_columms = ["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"]

student_mat_por = pd.merge(student_mat, student_por, on=merge_columms
                                                  , how="outer"
                                                  , suffixes=("_mat", "_por"))
```

## **üßπ Removing Invalid Values (Lo·∫°i b·ªè d·ªØ li·ªáu kh√¥ng h·ª£p l·ªá)**


```python
# G·ªôp ƒëi·ªÉm ƒë·ªÉ l√†m target
student_mat_por["G3"] = student_mat_por[["G3_mat", "G3_por"]].mean(axis=1)
student_mat_por["G2"] = student_mat_por[["G2_mat", "G2_por"]].mean(axis=1)
student_mat_por["G1"] = student_mat_por[["G1_mat", "G1_por"]].mean(axis=1)

student_mat_por = student_mat_por.drop(columns=["G1_mat","G1_por","G2_mat","G2_por", "G3_mat","G3_por"])
```

```python
order_col_list_cat = ["guardian", "schoolsup", "famsup", "paid", "activities", "higher", "romantic"]
order_col_list_num = ["traveltime", "studytime", "failures", "famrel", "goout", "Dalc", "Walc", "health", "absences", "freetime"]

# G·ªôp c√°c c·ªôt ph√¢n lo·∫°i
def merge_multiple_categorical(df, col_list):
    for col in col_list:
        df[col] = df[f"{col}_mat"].combine_first(df[f"{col}_por"])
        df = df.drop(columns=[f"{col}_mat", f"{col}_por"])
    return df

# G·ªôp c√°c c·ªôt s·ªë
def merge_multiple_numerical(df, col_list):
    for col in col_list:
      if col != "absences":
        df[col] = df[[f"{col}_mat", f"{col}_por"]].mean(axis=1)
      else:
        df[col] =  df[[f"{col}_mat", f"{col}_por"]].max(axis=1)
      df = df.drop(columns=[f"{col}_mat", f"{col}_por"])
    return df

student_mat_por = merge_multiple_categorical(student_mat_por, order_col_list_cat)
student_mat_por = merge_multiple_numerical(student_mat_por, order_col_list_num)
```
